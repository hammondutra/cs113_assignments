```python
# Exercise 2.8.1

# Defining both matrices
A = matrix(QQ, [[1, 3, -1],
                [2, 0, 4]])

U = matrix(QQ, [[1, 0, 2],
                [0, 1, -1]])
```

```python
# Finding the four fundamental subspaces of A

# Reducing A to RREF to find the column space of A
A, A.rref()
```

```python
"""
When we turn A into rref, we see that it has two pivot columns. This means that the columns are 
linearly independent and both contribute to the span of the column space. So, the basis of the 
column space is {(1, 2), (3, 0)}. The dimension of the column space is therefore 2.

dim(col(A)) = 2
"""
```

```python
# To get the basis and dimension of the null space, we can solve the homogeneous equation Ax = 0
A.augment(vector([0, 0]), subdivide=True).rref()
```

```python
"""
We have one non-pivot column, so this means that there are parametric solutions.

Let z = t:
The solutions to the system would look like (x = -2t, y = t, z = t) or t*(-2, 1, 1)

This means that the basis is a set with one vector (-2, 1, 1) and that the dimension of the null space is 1.

dim(N(A)) = 1
"""
```

```python
# To get the row space, we can look at the rref of the transpose of A to see if they are linearly independent
A.transpose(), A.transpose().rref()
```

```python
"""
The first and second rows are linearly independent and thus the basis for the row space will contain the first two columns of A^T
The basis for the column space is {(1, 3, -1), (2, 0, 4)}. The dimension of the row space is 2.

dim(row(A)) = 2
"""
```

```python
"""
Finally, for the left-null space, we can solve the homogeneous equation A^t * x = 0

As we can see from the matrix above:
[1 0]
[0 1]
[0 0]

we have a total of 0 non-pivot columns. This means that the basis for the left-null space is {} and
that dim(N(A^T)) = 0
"""
```

```python
# Finding the four fundamental subspaces of U

# Reducing U to RREF to find the column space of U
U, U.rref()
```

```python
"""
Analyzing it, we see that it has two pivot columns. This means that the first two columns are
linearly independent and both contribute to the span of the column space. 
So, the basis of the column space is {(1, 0), (0, 1)}.

The dimension of the column space is therefore 2.

dim(col(U)) = 2
"""
```

```python
# To get the basis and dimension of the null space, we can solve the homogeneous equation Ux = 0
U.augment(vector([0, 0]), subdivide=True).rref()
```

```python
"""
We have one non-pivot column, so this means that there are parametric solutions.

Let z = t:
The solutions to the system would look like (x = -2t, y = t, z = t) or t*(-2, 1, 1)

This means that the basis is a set with one vector (-2, 1, 1) and that the dimension of the null space is 1.

dim(N(U)) = 1
"""
```

```python
# To get the row space, we can look at the rref of the transpose of U to see if they are linearly independent
U.transpose(), U.transpose().rref()
```

```python
"""
The first and second rows are linearly independent and thus the basis for the row space will contain the first two rows of U^T
The basis for the column space is {(1, 0, 0), (0, 1, 0)}. The dimension of the row space is 2.

dim(row(U)) = 2
"""
```

```python
"""
Finally, for the left-null space, we can solve the homogeneous equation U^t * x = 0

As we can see from the matrix above:
[1 0]
[0 1]
[0 0]

we have a total of 0 non-pivot columns. This means that the basis for the left-null space is {} and
that dim(N(U^T)) = 0
"""
```

```python
# Exercise 2.8.3 (a)

A = matrix(QQ, [
    [1, 0, 0],
    [1, 15, 225],
    [1, 30, 900],
    [1, 45, 2025]
])

A, A.rref()
```

```python
# Showing that the system Ax = b is inconsistent

b = vector(QQ, [1, 3, 2, 2])

# 1. Create the augmented matrix
M = A.augment(b, subdivide=True)

# 2. Compute its RREF
M_rref = M.rref()
M_rref
```

```python
# Finding x_hat

A_t = A.transpose()
print(f"A transpose (A_t):\n{A_t}\n")

A_tA = A_t * A
print(f"A_t * A:\n{A_tA}\n")

A_tb = A_t * b
print(f"A_t * b:\n{A_tb}\n")

# Solving the equation (A_t*A) * x_hat = (A_t*b)
x_hat = A_tA.solve_right(A_tb)
print(f"x_hat is: {x_hat}\n")
```

```python
# Plotting the data points
D = [(0, 1), (15, 3), (30, 2), (45, 2)]

s = scatter_plot(D)

s.show()
```

```python
#Solve for x_hat
var("m, t, b")

model_line(t) = m*t + b # Linear model
fit_equations = find_fit(D, model_line)
print(fit_equations)

m_fit = fit_equations[1].rhs()
b_fit = fit_equations[0].rhs()

regression_line(t) = m_fit*t + b_fit

p_line = plot(regression_line, (t, 0, 50), color="purple", thickness=2)

show_all = s + p_line

show_all.show()
```

```python
# Finding the quadratic approximation equation
C_hat = x_hat[0]
D_hat = x_hat[1]
E_hat = x_hat[2]

parabola(t) = C_hat + D_hat*t + E_hat*t^2
print(f"Quadratic Model: y = {parabola(t).full_simplify()}")
```

```python
# Plotting both the linear and quadratic approximations together
p_quadratic = plot(parabola, (t, 0, 50), color="green", thickness=2)

show_all = s + p_line + p_quadratic

show_all.show()
```

```python
# Calculating error for linear model:
total_error_line = 0
for p in D:
    t_i = p[0]
    b_i = p[1]
    
    predicted_val = regression_line(t_i)
    
    # Calculate the squared error for the point
    error_sq = (b_i - predicted_val)^2
    
    total_error_line += error_sq

print(f"Total Sum of Squared Errors (SSE) (rounded to 5 decimal places): {total_error_line.n(digits=5)}")
```

```python
# Calculating error for quadratic model

total_error_parabola = 0
for p in D:
    t_i = p[0]
    b_i = p[1]
    
    predicted_val = parabola(t_i)
    
    # Calculate the squared error for the point
    error_sq = (b_i - predicted_val)^2
    
    total_error_parabola += error_sq

print(f"Total Sum of Squared Errors (SSE) (rounded to 5 decimal places): {total_error_parabola.n(digits=5)}")
```
