```sage
# 3.9.4 (c)

A = matrix(QQ, [[0, 1, 1],
                [1, -1, -1],
                [1, 0, 0]])


v1 = vector(QQ, [0, 0, 2])
v2 = vector(QQ, [1, 1, 3])
v3 = vector(QQ, [1, -1, -5])

S = matrix(QQ, [v1, v2, v3]).transpose()
print(f"S is:\n{S}\n")

print(f"S.rref() is: \n{S.rref()}")

# Multiplying S by A will give us the transformed basis vectors
transformed_basis_still_in_E = A*S # M is on the left!

transformed_basis_still_in_E

```

```sage
"""
Now, we have transformed each of our basis vectors, but the transformed vectors are still represented in the standard basis. So we have to convert them to the S basis.
We can do that by multiplying the transformed basis vectors by the change of basis matrix Q that takes us from E to S. This change of basis is the inverse of the matrix P that
takes us from S to E. Let's first find the matrix P
"""
P = S

# Now, we can take the inverse of P to find Q
Q = P^(-1)
Q
```

```sage
"""
Now that we have the change of basis matrix Q that takes us from E to S, we can change the basis of our transformed basis vectors from E to S. We can do this by multiplying 
Q by transformed_basis_still_in_E
"""
B = Q * transformed_basis_still_in_E
B
```
