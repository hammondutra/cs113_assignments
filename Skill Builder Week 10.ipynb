```python
v1 = vector(QQ, [2, 1])
v2 = vector(QQ, [0, 1])

s1 = v1/v1.norm()
print(s1)

a1 = v2 - (v2*s1)*s1
print(a1)

s2 = a1 / a1.norm()
print(s2)
```

```python
P = matrix([s1, s2]).transpose()
print(P)

P_inverse = P^(-1)
print(P_inverse)

B = matrix(QQ, [[2, 0], [0, -1]])

A = P*B*P^(-1)
A
```

```python
L_1 = 2
L_2 = -1

X_1 = A - L_1 * identity_matrix(2)
print(det(X_1))

X_2 = A - L_2 * identity_matrix(2)
print(det(X_2))
```

```python
# Multiplying X_1 * v1 (equivalent to applying the transformation to v1)
v1 = vector(QQ, [2, 1])

print(X_1*v1)

# Multiplying X_2 * v2 (equivalent to applying the transformation to v1)
v2 = vector(QQ, [1, -2])

print(X_2*v2)
```
