```sage
# (2.5.3) Applying the Gram-Schmidt Algorithm to the vector basis

v1 = vector(QQ, [ 1, 1, 0])
v2 = vector(QQ, [2, -2, 1])
v3 = vector(QQ, [0, 1, -1])

# Right now, A is not orthogonal, so we have to use Gran-Schmidt

# Normalize first vector
q1 = v1/v1.norm()

print(q1)

# Orthogonalize the second vector
a2 = v2 - v2.dot_product(q1)*q1
print(v2.dot_product(q1)*q1)
print(a2)

# Normalize the second vector
q2 = a2/a2.norm()
print(q2)

# Orthogonalize the third vector
a3 = v3 - v3.dot_product(q1)*q1 - v3.dot_product(q2)*q2
print(v3.dot_product(q1)*q1)
print(v3.dot_product(q2)*q2)

print(a3)

# Normalize the third vector
q3 = a3/a3.norm()
print(a3.norm())
print(q3)
```

```sage
# Verifying if the vectors are orthogonal to each other analytically

print(q1*q2)
print(q1*q3)
print(q2*q3)
```
