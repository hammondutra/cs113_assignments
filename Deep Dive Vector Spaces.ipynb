```python
# Compressing the image for various values of epsilon:

def compress_image(image_matrix, basis_matrix_row_inv, basis_matrix_row, epsilon):
    """
    Compresses an 8x8 image by applying a threshold to the
    S-basis coordinates of each row.
    """
    # Get S-basis coordinates for each row
    Img_Coords_S = image_matrix * basis_matrix_row_inv

    # Apply threshold
    Compressed_Coords_S = matrix(QQ, Img_Coords_S)
    zeros_created = 0
    for i in range(Compressed_Coords_S.nrows()):
        for j in range(Compressed_Coords_S.ncols()):
            if abs(Compressed_Coords_S[i, j]) < epsilon:
                Compressed_Coords_S[i, j] = 0
                zeros_created += 1

    # Convert back to standard basis
    Compressed_Img = Compressed_Coords_S * basis_matrix_row

    print(f"Sparsity: Created {zeros_created} zeros out of 64 coordinates ({zeros_created / 64.0 * 100.0:.1f}%)")
    
    return Compressed_Img

print("Compressed Image (Epsilon = 30):")
Img_30 = compress_image(Img, M_inv, M, 30)
matrix_plot(Img_30, cmap="gray", vmin=0, vmax=256)
```
